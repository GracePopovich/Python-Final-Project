```python
import seaborn as sns
```

```python
mpg=sns.load_dataset('mpg')
```

```python
mpg.head(1)
```

```python
#Create a visual showing the relationship between horsepower and fuel economy
sns.relplot(data=mpg, x="horsepower",y="mpg")
```

```python
#as horsepower increases,miles per gallon decreases.
```

```python
#Create a table summarizing fuel economy by country of origin
sns.barplot(data=mpg, x="origin",y="mpg")
```

```python
#Cars with an origin from Japan tend to have the best fuel efficiency.
#Cars with an origin from the USA tend to had the poorest fuel efficency.
```

```python
#Create a visual showing how fuel economy has changed over time
sns.lmplot(data=mpg, x="model_year",y="mpg")
```

```python
#Generally, cars made later have higher miles per gallon. 
```

```python
# How the number of cylinders affects miles per gallon.
sns.barplot(data=mpg, x="cylinders",y="mpg")
```

```python
#cars with 4 cylinders are on average the most fuel efficient.
#cars with 8 cylinders are on average the least fuel efficient.
```

```python
sns.relplot(data=mpg,x="horsepower", y="mpg", hue="weight")
```

```python
#generally, as cars get heavier, the horsepower on the vehicle goes up, and the miles per gallon goes down.
```
